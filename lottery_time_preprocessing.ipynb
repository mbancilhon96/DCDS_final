```python
data2=pd.merge(data_csv,batch_results, left_on="postId", right_on="Answer.surveycode",how="inner" )
```

# Remove users for incomplete experiments

```python
data2=data2.dropna(subset=['time_end_experiment','age']).reset_index()
len(data2)
```

# Create sheets dataframe

```python
data['sheet12'][1]
```

```python
postId=[]
p_list=[]
estimate=[]
estimateVal=[]
time=[]
lottery_list=[]
x1=[]
x2=[]
ev_list=[]
rrp_list=[]
sheet_no=[]
age=[]
ev=[]
gender=[]
education=[]

for i in range(25):
    sheet="sheet"+str(i)
    for index,item in enumerate(data2[sheet]):
        if str(item)=="nan":
            print("postId ", data2['postId'][index]," sheet no. ",sheet)
            pass
        else:
            item= json.loads(item)
#             print(item)
            postId.append(data2['postId'][index])
            age.append(data2['age'][index])
            gender.append(data2['gender'][index])
            education.append(data2['education'][index])
            p=float(item['p'])
            p_list.append(p) 
            estimate.append(item['estimate']) 
            try: estimateVal.append(item['estimateVal'])
            except: estimateVal.append("null")
            time.append(item['time'])
            x1.append(item['x1'])
            x2.append(item['x2'])
            lotterychoice= item['lotterychoice']
            lottery_list.append(lotterychoice)
            #calculate ev
            ev= (p*float(item['x1'])) + ((1-p)*float(item['x2']))
            ev_list.append(ev)
            #calculate ce
            payments=item['payments']
            letter=lotterychoice.rsplit()
            if lotterychoice[-1]=="a":
                ce= (payments[int(lotterychoice[:-1])-1]+ payments[int(lotterychoice[:-1])])/2
            if lotterychoice[-1]=="b":
                ce= (payments[int(lotterychoice[:-1])-1]+ payments[int(lotterychoice[:-1])])/2
            #calculate rrp
            rrp= (ev-ce)/ev
            rrp_list.append(rrp)
            #append sheet number
            sheet_no.append(sheet)            
```

```python
d = {'postId': postId, 'age':age, 'gender':gender,'education':education,'estimate':estimate,'estimateVal':estimateVal,'rrp': rrp_list, 'p':p_list,'time':time,'sheet_no':sheet_no,'x1':x1,'x2':x2,'lotterychoice':lottery_list}
df = pd.DataFrame(data=d)
df.to_csv("lottery_time.csv")
```
